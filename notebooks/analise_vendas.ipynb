1 - Importar libs e dados

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/vendas.csv")
df
```

2 - Faturamento

```python
df["faturamento"] = df["preco"] * df["quantidade"]
df
```

3 - KPIs principais

```python
df[["faturamento"]].sum()
df.groupby("categoria")["faturamento"].sum()
df.groupby("regiao")["faturamento"].sum()
```

4 - Gráfico

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data=df,
    x="regiao",
    y="faturamento",
    estimator=sum
)
plt.title("Faturamento por Região")
plt.show()
```
